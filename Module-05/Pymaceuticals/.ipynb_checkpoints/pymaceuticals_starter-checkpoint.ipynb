```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
from scipy.stats import linregress

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_table = pd.merge(clinical_trial_df,mouse_drug_df, on= 'Mouse ID', how ='left')

# Display the data table for preview
merge_table.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_time_group = merge_table.groupby(['Drug', 'Timepoint'])
mean_tumor_volume = drug_time_group['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
drug_time_df = pd.DataFrame(mean_tumor_volume).reset_index()

# Preview DataFrame
drug_time_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
#standard_error_volume = sem(mean_tumor_volume)
standard_error_volume = drug_time_group['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
standard_error_volume_df = pd.DataFrame(standard_error_volume).reset_index()

# Preview DataFrame
standard_error_volume_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
drug_time_munging = drug_time_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
drug_time_munging.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
drugs = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
colors = ['red', 'blue', 'green', 'black']
markers = ['o', '^', 's', 'd']

#plots a single drug, sterr is True when error bars are being plotted(requires sdf), otherwise set to False
def plot_drug(df, sdf, drug,x, y, color, marker,ax,sterr = True):
    x_axis = df.loc[df['Drug'] == drug, x]
    y_axis = df.loc[df['Drug'] == drug, y]
    if sterr != True:
        ax.errorbar(x_axis, y_axis, linewidth=1, linestyle = ':', marker=marker, color = color)
    else:
        error = sdf.loc[df['Drug'] == drug, y]
        ax.errorbar(x_axis, y_axis, error, linewidth=1, linestyle = ':', marker=marker, color = color)
        
#plots multiple drugs        
def create_plot(df,sdf, drugs, x, y, colors, markers, ax, sterr = True):
    for i in range(len(drugs)):
        plot_drug(df, sdf, drugs[i], x, y,colors[i], markers[i],ax,sterr)
        
#labels plot        
def label_plot(title, x_label, y_label, ax, label= drugs):
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    plt.legend(labels = label,loc="best", fontsize="small", fancybox=True)
    ax.grid()
        
create_plot(drug_time_df, standard_error_volume_df, drugs,'Timepoint', 'Tumor Volume (mm3)', colors, markers, ax)   
label_plot("Tumor Response to Treatment", "Time (Days)", "Tumor Volume (mm3)", ax)

# Save the Figure
plt.savefig('data/Tumor_Response.png')
fig.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site = drug_time_group['Metastatic Sites'].mean()

# Convert to DataFrame
mean_met_site_df = pd.DataFrame(mean_met_site)

# Preview DataFrame
reset_mean_met_site_df = mean_met_site_df.reset_index()#used for plot
mean_met_site_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
standard_error_met = drug_time_group['Metastatic Sites'].sem()

# Convert to DataFrame
standard_error_met_df = pd.DataFrame(standard_error_met)

# Preview DataFrame
reset_standard_error_met_df = standard_error_met_df.reset_index()#used for plot
standard_error_met_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_site_munging = reset_mean_met_site_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Metastatic Sites')

# Preview that Reformatting worked
mean_met_site_munging.head()
```

```python
# Generate the Plot (with Error Bars)
fig_2, ax_2 = plt.subplots()

create_plot(reset_mean_met_site_df, reset_standard_error_met_df, drugs, 'Timepoint', 'Metastatic Sites', colors, markers, ax_2)
label_plot("Metastatic Spread During Treatment", "Treatment Duration (Days)", "Met. Sites", ax_2)

# Save the Figure
plt.savefig('data/Metastatic_Spread.png')

# Show the Figure
fig_2.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = drug_time_group['Mouse ID'].count()

# Convert to DataFrame
mouse_count_df = pd.DataFrame(mouse_count).reset_index()
rename_mouse_count_df = mouse_count_df.rename(columns= {'Mouse ID': 'Mouse Count'})
# Preview DataFrame
#mouse_count_df.head()
rename_mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_munging = rename_mouse_count_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Mouse Count')

# Preview the Data Frame
mouse_count_munging.head()
```

```python
# Generate the Plot (Accounting for percentages)
fig_3, ax_3 = plt.subplots()
#total_mouse = rename_mouse_count_df.loc[rename_mouse_count_df['Timepoint'] == 0,'Mouse Count']

#Get the starting mouse count for JUST the 4 drugs. NOT Dynamic!!
mouse_count_df['percent_mouse_count'] = (rename_mouse_count_df['Mouse Count']/rename_mouse_count_df['Mouse Count'][0])*100    

create_plot(mouse_count_df, 'none', drugs, 'Timepoint', 'percent_mouse_count', colors, markers, ax_3,sterr = False)
label_plot("Survival During Treatment", "Time (Days)", "Survival Rate (%)", ax_3)

# Save the Figure
plt.savefig('data/Survival_Rate.png')

# Show the Figure
fig_3.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
def calculate_percent_change(drug):
    percent_change = drug_time_df.loc[drug_time_df['Drug'] == drug, "Tumor Volume (mm3)"]
    percent_change =  (percent_change.iloc[-1]/percent_change.iloc[0]) *100
    percent_change -= 100
    return percent_change

#create dictionary to hold percent change per drug store in dataframe
drug_percent_change = {}
for drug in drug_time_df['Drug'].unique():
    drug_percent_change[drug]=  calculate_percent_change(drug)
percent_change_df = pd.DataFrame([drug_percent_change]).T.reset_index()       
    
# Display the data to confirm
rename_percent_change_df = percent_change_df.rename(columns= {'index': 'Drug', 0:'Percent Change'})

rename_percent_change_df
```

```python
# Store all Relevant Percent Changes into a Tuple
fig_4, ax_4 = plt.subplots()
percent_change_x_axis = ['Capomulin', 'Infubinol','Ketapril','Placebo']

percent_change_Cap = rename_percent_change_df.loc[rename_percent_change_df['Drug'] == 'Capomulin', 'Percent Change'].mean()
percent_change_Inf = rename_percent_change_df.loc[rename_percent_change_df['Drug'] == 'Infubinol', 'Percent Change'].mean()
percent_change_Ket = rename_percent_change_df.loc[rename_percent_change_df['Drug'] == 'Ketapril', 'Percent Change'].mean()
percent_change_Pla = rename_percent_change_df.loc[rename_percent_change_df['Drug'] == 'Placebo', 'Percent Change'].mean()

percent_change_tuple = (percent_change_Cap, percent_change_Inf, percent_change_Ket, percent_change_Pla)

# Splice the data between passing and failing drugs
for i in range(len(percent_change_tuple)):
    if percent_change_tuple[i] > 0:
        ax_4.bar(percent_change_x_axis[i], percent_change_tuple[i], color = 'red', align = 'edge', width = -1)
    else:
        ax_4.bar(percent_change_x_axis[i], percent_change_tuple[i], color = 'green', align = 'edge', width = -1)

# Orient widths. Add labels, tick marks, etc. 
ax_4.set_title("Tumor Change Over 45 Day Treatment")
#ax_4.set_xlabel("Drugs")
ax_4.set_ylabel("% Tumor Volume Change")
plt.yticks(np.arange(-20, 65, 20))
plt.ylim(-30,70)
plt.grid()

# Use functions to label the percentages of changes 
#autolabel() ref:https://matplotlib.org/1.5.1/examples/api/barchart_demo.html
def autolabel(rects):
    
    for i in range(len(rects)):
        height = rects[i].get_height()
        x = rects[i].get_x() + rects[i].get_width()/2
        if height > 0:
            ax_4.text(x=x, y=3, s=f"{height:.0f}%", color='white', ha='center', va='bottom')
        else:
            ax_4.text(x=x, y=-7, s=f"-{height:+.0f}%", color='white', ha='center', va='bottom')
            
# Call functions to implement the function calls
autolabel(ax_4.patches)

# Save the Figure
plt.savefig('data/Tumor_Change_45_Days.png')

# Show the Figure
fig_4.show()
```

```python
#3 Observable Trends
#1. Of the four drugs Capomulin has the highest survival rate, nearly double that of the other drugs.
#2. Ketapril seems to have very little effect, having about the same performance as the Placebo.
#3. Infubinol has the worst performance as it has the lowest survival rate, despite lower metastatic spread than the Placebo and Ketapril.
#4. It seems the survival rate is more impacted by the tumor volume than the metastatic spread, since the most effective drug, Capomulin has both a dramatic reduction in volume, and a high survival rate.
#   While the other three were very close in both their high tumor growth and their low survival rates, but differed in their met. site spread. 
```
